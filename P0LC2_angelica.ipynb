Section 2 (Pandas)

```python
import pandas as pd
from datetime import datetime
```

```python
df = client.query('''
SELECT *
FROM `ftds-hacktiv8-project.phase0_ftds_045_rmt.taxi_fare`
''').to_dataframe()
```

```python
df['trip_start_timestamp'] = pd.to_datetime(df['trip_start_timestamp'])
df['Jam Mulai'] = df['trip_start_timestamp'].dt.hour
period_list = []
for jam in df['Jam Mulai']:
    if 4 <= jam <= 9:
        period_list.append('morning')
    elif 10 <= jam <= 15:
        period_list.append('daytime')
    elif 16 <= jam <= 18:
        period_list.append('evening')
    elif 19 <= jam <= 23:
        period_list.append('night')
    else:
        period_list.append('midnight')
df['period'] = period_list
```

```python
total_jarak = df[
    (df['company'] == 'Federway Int.') &
    (df['trip_start_timestamp'].dt.year == 2022)
]['trip_km'].sum()

print(f"Total jarak Federway Int. (2022): {total_jarak} km")
```
